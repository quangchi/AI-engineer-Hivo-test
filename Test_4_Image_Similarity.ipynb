```python
from image_processor import ImageProcessor
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_images(processed_image1, processed_image2, similarity):
    # Plot the two images side by side
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # Create 1 row and 2 columns
    
    # Display the first image
    axes[0].imshow(processed_image1)
    axes[0].axis('off')  # Turn off axis
    axes[0].set_title("processed_image1")
    
    # Display the second image
    axes[1].imshow(processed_image2)
    axes[1].axis('off')  # Turn off axis
    axes[1].set_title("processed_image2")
    plt.suptitle("similarity: {}".format(similarity), fontsize=20, fontweight='bold')
```

```python
processor = ImageProcessor()
```

# Test 4.1: Image Similarity, Calculate similarity score between (<span style="color:red">test_image, test_image</span>) 

```python
with open('images/demo/test_image.jpg', 'rb') as f:
    image_data = f.read()
processed_image1 = processor.preprocess_image(image_data)
with open('images/demo/test_image.jpg', 'rb') as f:
    image_data = f.read()
processed_image2 = processor.preprocess_image(image_data)
similarity = processor.calculate_image_similarity(processed_image1, processed_image2)
assert 0 <= similarity <= 1
assert abs(similarity - 1.0) < 1e-6  # Should be 1.0 for identical images
```

```python
similarity
```

```python
plot_images(processed_image1, processed_image2, similarity)
```
