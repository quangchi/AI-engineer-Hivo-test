```python
from image_processor import ImageProcessor
import numpy as np
import matplotlib.pyplot as plt
```

```python
processor = ImageProcessor()
```

# Test 3: Tag Generation

```python
with open('images/demo/test_image.jpg', 'rb') as f:
    image_data = f.read()
processed_image = processor.preprocess_image(image_data)
tags = processor.generate_tags(processed_image)
assert len(tags) >= 5
assert all(isinstance(tag, tuple) and len(tag) == 2 for tag in tags)
assert all(0 <= score <= 1 for _, score in tags)
```

```python
tags
```

```python
plt.figure(figsize=(6, 6))  # Adjust the size of the figure if necessary
plt.imshow(processed_image)
plt.axis('off')  # Hide axes for a cleaner look
plt.show()
```
